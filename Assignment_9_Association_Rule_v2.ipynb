```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("book.csv")
df.head() 
```

# Apriori Algorithm

**1. Association rules with 7% Support and 70% confidence**

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_items=apriori(df, min_support=0.07, max_len=3,use_colnames = True)
frequent_items
```

```python
frequent_items.sort_values('support',ascending=False,inplace=True)
frequent_items
```

```python
rules=association_rules(frequent_items,metric='confidence', min_threshold=0.7)
rules
# rules[(rules['confidence']>0.7)]
```

```python
rules.sort_values('confidence', ascending=False)
```

```python
len(rules)
```

```python
# visualization of obtained rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

```python
plt.bar(x = list(range(1,11)),height = frequent_items.support[1:11],color='rgmyk');
```

```python
plt.xticks(list(range(1,11)),frequent_items.itemsets[1:11])
plt.xlabel('item-sets');plt.ylabel('support')
```

# Using Apyori package (optional)

```python
!pip install apyori
from apyori import apriori
import numpy as np
```

```python
df = pd.read_csv('book.csv')
```

```python
def change_value(column):
  df[column].replace(1, column, inplace=True)
```

```python
for i in df.columns:
  change_value(i)
```

```python
df = df.replace(0, np.nan)
```

```python
df
```

```python
df.shape
```

```python
#Convert Pandas DataFrame into a list of lists
records = []
for i in range (0, 2000):
  records.append([str(df.values[i,j]) for j in range(0,11)])
```

```python
association_rules = apriori(records,min_support=0.01, min_confidence=0.4, min_lift=3, max_length=3)
```

```python
association_results = list(association_rules)
```

```python
len(association_results)
```

```python
for i in range(0, len(association_results)):
  print(association_results[i][0])
```

```python
for item in association_results:
    # first index of the inner list
    # Contains base item and add item
    pair = item[0]
    items = [x for x in pair]
    print("Rule: With " + items[0] + " you can also read " + items[1])

    # second index of the inner list
    print("Support: " + str(item[1]))

    # third index of the list located at 0th
    # of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
